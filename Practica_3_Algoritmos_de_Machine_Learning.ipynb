### Creando un Decision Tree

Aqu hay un ejemplo de un clasificador de rbol de decisin en `scikit-learn`. Comenzaremos por definir algunos datos etiquetados bidimensionales:

### Decision Trees y Over-fitting

Un problema con los rboles de decisin es que es muy fcil crear rboles que ** se ajustan en exceso ** a los datos. Nos encontramos con **overfitting**. Es decir, son lo suficientemente flexibles como para que puedan aprender la estructura del ruido en los datos en lugar de la seal! Por ejemplo, observe dos rboles construidos en dos subconjuntos de este conjunto de datos:

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()

plt.figure()
visualize_tree(clf, X[:200], y[:200], boundaries=False)
plt.figure()
visualize_tree(clf, X[-200:], y[-200:], boundaries=False)
```

Los detalles de las clasificaciones son completamente diferentes! Eso es una indicacin de **ajuste excesivo**: cuando predice el valor para un nuevo punto, el resultado refleja ms el ruido en el modelo que la seal.

## Ensembles of Estimators: Random Forests

Una posible forma de abordar el **overfitting** es usar un **ensembles method**: este es un meta-estimador que esencialmente promedia los resultados de muchos estimadores individuales que sobre-ajustan los datos. Sorprendentemente, las estimaciones resultantes son mucho ms slidas y precisas que las estimaciones individuales que las componen.

Uno de los mtodos de conjunto ms comunes es el **Random Forest**, en el que el conjunto est formado por muchos rboles de decisin que de alguna manera estn perturbados.

Hay volmenes de teora y precedentes sobre cmo aleatorizar estos rboles, pero como ejemplo, imaginemos que un conjunto de estimadores se ajusta a los subconjuntos de los datos. Podemos tener una idea de cmo se veran estos de la siguiente manera:

```python
def fit_randomized_tree(random_state=0):
    X, y = make_blobs(n_samples=300, centers=4,
                      random_state=0, cluster_std=2.0)
    clf = DecisionTreeClassifier(max_depth=15)
    
    rng = np.random.RandomState(random_state)
    i = np.arange(len(y))
    rng.shuffle(i)
    visualize_tree(clf, X[i[:250]], y[i[:250]], boundaries=False,
                   xlim=(X[:, 0].min(), X[:, 0].max()),
                   ylim=(X[:, 1].min(), X[:, 1].max()))
    
from ipywidgets import interact
interact(fit_randomized_tree, random_state=(0, 100));
```

Vea cmo cambian los detalles del modelo en funcin de la muestra, mientras que las caractersticas ms grandes siguen siendo las mismas.

El clasificador de bosque aleatorio har algo similar a esto, pero usa una versin combinada de todos estos rboles para llegar a una respuesta final:

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=0)
visualize_tree(clf, X, y, boundaries=False);
```

Al promediar **100 modelos perturbados al azar**, terminamos con un modelo general que se ajusta mucho mejor a nuestros datos.

**(Nota: anteriormente, aleatorizamos el modelo mediante submuestreo ... 
Los bosques aleatorios utilizan medios de aleatorizacin ms sofisticados, sobre los cuales puede leer, por ejemplo, la [documentacin de scikit-learn](http://scikit-learn.org/stable/modules/ensemble.html#forest))**

## Decision Trees y Random Forest para Classificacin de Dgitos

Tomando el dataset de **hand-written digits** vamos a probar la eficacia de un clasificador por Decision Tree.

Cargamos el dataset en memoria.

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.keys()
```

```python
X = digits.data
y = digits.target
print(X.shape)
print(y.shape)
```

Para recordarnos lo que estamos viendo, visualizaremos los primeros puntos de datos:

```python
# set up the figure
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    
    # label the image with the target value
    ax.text(0, 7, str(digits.target[i]))
```

Podemos clasificar rpidamente los dgitos utilizando un rbol de decisin de la siguiente manera:

Hacemos una particin de los datos en set de entrenamiento y set de testeo con la funcin `train_test_split`: 

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

Definimos el `clasificador` y lo ajustamos a nuestros datos de entrenamiento.

```python
clf = DecisionTreeClassifier(max_depth=5)
clf.fit(Xtrain, ytrain)
```

Predecimos las etiquetas del test dataset, para posteriormente poder comprobar su validez contra las reales:

```python
ypred = clf.predict(Xtest)
```

Podemos comprobar la exactitud de este clasificador:

```python
print('Precisin sobre el test dataset: ', metrics.accuracy_score(ypred, ytest)*100, '%')
```

### Matriz de confusin

Visualicemos el comportamiento de nuestro rbol clasificador con la matriz de confusin.

Podis encontrar detalles sobre esta matriz en [este artculo](https://data-speaks.luca-d3.com/2018/01/ML-a-tu-alcance-matriz-confusion.html).

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(ytest, ypred)
print(classification_report(ytest, ypred))
print(cm)
```

```python
sns.heatmap(cm, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');
```

### Pregunta 2 (6 puntos):
### Repita esta clasificacin con ``sklearn.ensemble.RandomForestClassifier``  Mejoran los resultados de precisin en el test dataset? Represente la matriz de correlacin.

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

digits = load_digits()
X = digits.data
y = digits.target
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
clf = RandomForestClassifier(random_state=0)
clf.fit(Xtrain, ytrain)
ypred = clf.predict(Xtest)
accuracy = metrics.accuracy_score(ypred, ytest)*100
print(accuracy)
print(classification_report(ytest, ypred))
cm = confusion_matrix(ytest, ypred)
sns.heatmap(cm, square=True, annot=True, cbar=False) 
plt.xlabel('predicted value')
plt.ylabel('true value');
```

Con una precisin del 98% con accuracy random forest es mayr al accuracy del decision tree, configurando un mejor modelo
